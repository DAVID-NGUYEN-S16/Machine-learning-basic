> Problem 1
+ A farmer has a total of 10 hectares (10 hectares) of cultivated land. He plans to grow coffee and pepper on this land with a total cost of no more than 16T (million VND). The cost to grow coffee is 2T for 1ha, to grow pepper is 1T/ha. Coffee growing time is 1 day/ha and pepper is 4 days/ha; while he only has a total of 32 days. After deducting all costs (including planting costs), each hectare of coffee yields a profit of 5T, each hectare of pepper yields a profit of 3T. Ask him how to grow to maximize profits? (The numbers may be absurd because they were chosen to give the problem a good solution.)

+ Data:
    - set up:
        + x is the number of simple ha a coffee
        + y is the number of simple ha a pepper
+ So, this some constract:
    - $x + 4*y \leq 32$
    - $2*x + y \leq 16$
    - $x + y \leq 10$
    - $x, y \geq 0$
    - $f(x, y) = argmax(5*x + 3*y)$ so this subjective function

```python
from cvxopt import matrix, solvers
```

```python
'''
So this, include cvxopt had compute in formular linear Programming standard
. That is find the solution for the function to reach the minimum value
'''
```

- So, constract will changes:
    - $x + 4*y \leq 32$
    - $2*x + y \leq 16$
    - $x + y \leq 10$
    - $-x, -y \leq 0$
    - $f(x, y) = argmin(-5*x - 3*y)$ so this subjective function

```python
c = matrix([-5., -3.])
G = matrix([
    [1., 4.],
    [2., 1.],
    [1., 1.],
    [-1., 0.],
    [0., -1.]
])
h = matrix([32., 16., 10., 0., 0.])
solvers.options['show_progress'] = False
sol = solvers.lp(c, G.T, h)
print('Solution"')
print(sol['x'])
```
