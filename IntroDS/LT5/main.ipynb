## 1.2 Trực quan hóa bằng biểu đồ histogram, plotbox


```python
import matplotlib.pyplot as  plt
```

```python
def show_hist(ls):
    name_col = ls.columns
    fig, axis = plt.subplots(int((len(name_col) + 1)/3), 3, figsize=(20, 30))
    fig.suptitle("OVERVIEW DATA", fontsize=16)
    axis = axis.flatten()
    print(axis.flatten())
    for i in range(len(name_col)):
        axis[i].hist(ls[name_col[i]])
        axis[i].set_title(name_col[i])
        
        
```

```python
show_hist(data)
```

```python
def show_boxplot(ls):
    name_col = ls.columns
    fig, axis = plt.subplots(int((len(name_col) + 1)/3), 3, figsize=(20, 30))
    fig.suptitle("OVERVIEW DATA", fontsize=16)
    axis = axis.flatten()
    print(axis.flatten())
    for i in range(len(name_col)):
        axis[i].boxplot(ls[name_col[i]])
        axis[i].set_title(name_col[i])
show_boxplot(data)
```

# Các nhìn biêu đồ boxplot thì

## 1.10 Tính mỗi liên quan giữa các đặc trưng khác với bệnh tiểu đường, cho biết đâu là đặc trưng có ảnh hưởng nhất tới bệnh tiểu đường?

```python
ls = []
```

```python
for name in data.columns:
    if 'Class variable' in name: continue
    corr, _ = pearsonr(data[name], data['# 9. Class variable (0 or 1)'])
    ls.append([name, corr])
```
