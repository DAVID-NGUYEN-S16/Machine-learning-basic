```python
import glob
```

```python
list_file = glob.glob('*.txt')
```

```python
ls_name = []
for path in list_file:
    with open(path, 'r') as f:
        s  = f.read().replace('\n', '') .split(' ')
        ls_name += s
        f.close()
    
output = open('output.txt', 'w')
for x in set(ls_name):
    cnt = ls_name.count(x)
    output.write(f'{x}: {cnt} \n')
output.close()
```

```python
## aViết 2 chương trình đệ quy và không đệ quy để in ra các số Fibonacci từ 1
# tới n với n nhập vào từ bàn phím
def fib_ko_de_quy(n):
    F = [0]*(n+1)
    F[1] = 1
    F[2] = 1
    for i in range(3, n+1):
        F[i] = F[i-1] + F[i-2]
    return F[n]
def fib_de_quy(n):
    if(n <= 2): return 1
    return fib_de_quy(n-1) + fib_de_quy(n-2);
print(fib_ko_de_quy(8))
print(fib_de_quy(8))
```

```python
#  Viết 2 chương trình đệ quy và không đệ quy tìm Ước số chung lớn nhất của
# 2 số a,b bất kỳ nhập vào từ bàn phím
def gcd_not(a, b):
    for i in reversed(range(min(a, b)+1)):
        if(a % i == 0 and b % i == 0):
            return i
    # return F[n]
def gcd(a, b):
    if b == 0: return a
    else :return gcd(b, a%b);
# print(fib_ko_de_quy(8))
print(gcd_not(8, 4))
print(gcd(8, 4))
```
